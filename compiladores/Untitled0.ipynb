```python

caminho_arquivo_Automato = r"/content/drive/MyDrive/compiladores/Automato.txt"

caminho_arquivo_fonte = r"/content/drive/MyDrive/compiladores/teste.txt"

estados =[]
simbolos = []
estadosFinais = []
regrasTransicao = []
estadoAtual = ''
variavel = ''

def inicializacao():
    global estados,simbolos,estadosFinais,regrasTransicao,estadoAtual
    arquivo = open(caminho_arquivo_Automato,'r')
    estados = arquivo.readline().strip().split(',')
    linha_simbolos = arquivo.readline().strip()
    for c in linha_simbolos:
        simbolos.append(c)
    estadosFinais = arquivo.readline().strip().split(',')
    linha = arquivo.readline().strip()
    while linha:
        regrasTransicao.append(linha)
        linha = arquivo.readline().strip()
    estadoAtual = estados[0]
    arquivo.close()

def exec_afd(linha): #Automato finito Deterministico
    global estados,simbolos,estadosFinais,regrasTransicao,estadoAtual   #define as variaveis armazenadas como globais
    for c in linha:
        for regra in regrasTransicao:
            r = regra.split(':')
            if r[0] == estadoAtual and (c in r[1]):
                estadoAtual = r[2]
                print(f'{estadoAtual} -> {c}')

    print(linha,"Terminou no estado",estadoAtual)
    flag = 0
    for ef in estadosFinais:
        if ef.split(':')[0] == estadoAtual:
            print("Reconhecido:",ef.split(":")[1])
            flag = 1
    if flag == 0:
        print("Linha não reconhecida")

    estadoAtual = estados[0]


def lerInput():
    arquivo = open(caminho_arquivo_fonte,'r')
    linhas = arquivo.readlines()
    for linha in linhas:
        exec_afd(linha.strip())


inicializacao()
lerInput()

```
